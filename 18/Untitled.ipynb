```python
# B22DCDT321 - Nguyá»…n Quang Trung - Class: E22CNPM02

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("students.csv")

students = df["student_name"]

web = df["web_programming"]
python_prog = df["python_programming"]
algo = df["algorithms_and_data_structures"]
os = df["operating_systems"]
db = df["database_management"]

plt.plot(students, web, marker='o', label="Web Programming")
plt.plot(students, python_prog, marker='o', label="Python Programming")
plt.plot(students, algo, marker='o', label="Algorithms & DS")
plt.plot(students, os, marker='o', label="Operating Systems")
plt.plot(students, db, marker='o', label="Database Management")

plt.title("Results")
plt.xlabel("Students")
plt.ylabel("Grade")
plt.grid(True)
plt.xticks(rotation=45)
print("# B22DCDT321 - Nguyá»…n Quang Trung - Class: E22CNPM02")
print()
plt.legend()
plt.show()

x = range(len(students))
bar_width = 0.15

plt.bar([i - 2*bar_width for i in x], web, width=bar_width, label="Web Programming")
plt.bar([i - bar_width for i in x], python_prog, width=bar_width, label="Python Programming")
plt.bar(x, algo, width=bar_width, label="Algorithms & DS")
plt.bar([i + bar_width for i in x], os, width=bar_width, label="Operating Systems")
plt.bar([i + 2*bar_width for i in x], db, width=bar_width, label="Database Management")

plt.title("Results - Bar Chart")
plt.xlabel("Students")
plt.ylabel("Grade")
plt.xticks(x, students, rotation=45)  
plt.legend()
plt.show()

```

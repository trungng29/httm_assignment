```python
# B22DCDT321 - Nguyễn Quang Trung - Class: E22CNPM02

import pandas as pd
import seaborn as sns

df = pd.read_csv("titanic.csv")

g = sns.catplot(
    x = "who",
    y = "survived",
    col = "class",
    hue="who",
    data = df,
    kind = "bar",
    errorbar = None,
    aspect = 1.0,
    palette={"man": "skyblue", "woman": "orange", "child": "gray"}, 
    legend = False
)

g.set_axis_labels("", "Survival Rate")

print("# B22DCDT321 - Nguyễn Quang Trung - Class: E22CNPM02")
print()
print(df)
```

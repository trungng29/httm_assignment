```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Data
names = np.array(['Ann', 'Joe', 'Mark'])
heights = np.array([1.5, 1.78, 1.6])
weights = np.array([65, 46, 59])

bmi = weights / (heights ** 2)

def classify_bmi(value):
    if value < 18.5:
        return "Underweight"
    elif value > 25:
        return "Overweight"
    else:
        return "Normal"

categories = np.array([classify_bmi(x) for x in bmi])

df = pd.DataFrame({
    "Names": names, 
    "Heights (m)": heights,
    "Weights (kg)": weights,
    "BMI": bmi,
    "Category": categories
})

print(df)
print()

# Encode nhãn BMI bằng sklearn
encoder = LabelEncoder()
df["Category_Code"] = encoder.fit_transform(df["Category"])
print(df)
print()

# Vẽ biểu đồ bằng matplotlib
plt.figure(figsize=(6, 4)) # Xem kĩ lại cú pháp và tham số của dòng này
plt.bar(df["Names"], df["BMI"], color="skyblue", edgecolor="black") # Xem kĩ lại cú pháp và tham số của dòng này
plt.axhline(18.5, color="green", linestyle="--", label="Lowerbound (18.5)") # Xem kĩ lại cú pháp và tham số của dòng này
plt.axhline(25, color="red", linestyle="--", label="Upperbound (25)") # Xem kĩ lại cú pháp và tham số của dòng này
plt.title("BMI of Individuals")
plt.ylabel("BMI")
plt.xlabel("Names")
plt.legend() # Xem kĩ lại cú pháp và tham số của dòng này
plt.show()

```

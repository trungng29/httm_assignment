```python
import numpy
from numpy import loadtxt
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import load_model


########### 1. Load dữ liệu và chia train, val và test ###########
dataset = loadtxt("diabetes.csv", delimiter=",")

# print(dataset)
x = dataset[:, 0:8]
y = dataset[:, 8]

x_train_val, x_test, y_train_val, y_test =  train_test_split(x, y, test_size=0.2)
# Phần nhiều bao giờ cũng dành cho train và val, phần ít dành cho test
x_train, x_val, y_train, y_val = train_test_split(x_train_val, y_train_val, test_size=0.2) 

########### 2. Xây dựng kiến trúc model ###########
model = Sequential()
model.add(Dense(16, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.summary()

########### 3. Compile model ###########
# Loss, Optimizer, Metric
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

########### 4. Train model ###########
# Epoch, Batch_size, Validation
model.fit(x_train, y_train, epochs=100, batch_size=8, validation_data=(x_val, y_val))

########### 5. Save/load model ###########
model.save("mymodel.h5")
model = load_model("mymodel.h5")
loss, acc = model.evaluate(x_test, y_test)
print("Loss: ", loss)
print("Accuracy: ", acc)

########### Thực hiện dự đoán ###########
x_new = x_test[100]
y_new = y_test[100]

x_new = numpy.expand_dims(x_new, axis=0) 

y_predict = model.predict(x_new)
print("Gia tri du doan = ", y_predict)
print("Gia tri dung la = ", y_new)
```
